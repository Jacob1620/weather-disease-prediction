### Data Description

The data set for this project was obtained from [Kaggle](https://www.kaggle.com/datasets/orvile/weather-related-disease-prediction-dataset?resource=download) and consists of:

- **Observations**: $5,200$
- **Variables**: $51$ Total, $0$ Missing
- **Predictors ($50$ total)**:
    - **Numerical**: _Age_, _Temperature_ ($^{\circ} \text{C}$), _Wind Speed_ ($\frac{\text{km}}{\text{h}}$), _Humidity_ ($\%$)
    - **Binary**: _Gender_ and $45$ Medical Symptoms (e.g., _Cough_, _Vomiting_, _Headache_)
- **Response Variable**: _Prognosis_ (e.g., "Arthritis", "Eczema", "Stroke", "Common Cold")

### Project Objectives

This project intends to:

- Analyze which of the hydro-meteorological features shape the prognosis of diseases.
- Use clinical symptom and hydro-climate data to accurately predict when people get sick.
- Establish which features contribute the most to accurate prediction.
- Fit prediction models with high sensitivity and precision as measured by the F1 score.

### Modeling Approaches

#### Regression-Based Models

We trained several regression-based classifiers:
- Linear Regression
- Multinomial Logistic Regression
- Penalized Regression (using $\alpha$):
    - Ridge Regression ($\alpha = 0$)
    - Elastic Net Regression ($\alpha = \frac{1}{2}$)
    - Lasso ($\alpha = 1$)
- Principal Components Regression

**Overall Accuracy**: $\sim 83.43 \%$ across all models.


**Key Predictors**:

- **Linear**: _sneezing_, _back pain_, _eye pain_.
- **Penalized**: _nausea_, _headache_, _dizziness_.

**Key Prognoses** (Multinomial): _Stroke_, _Migraine_, _Influenza_

#### Neural Networks

We implemented a neural network model to classify patient prognosis. The most accurate model used 3 hidden layers with ReLU activation functions and L2 regularization, then a final layer to output softmax probabilities. It was trained over 100 epochs with early stopping, batch size of 128, Adam optimizer, and the sparse categorical cross-entropy loss function.

**Accuracy**: $\sim 95.7 \%$

We demonstrate the strong ability of the model to identify complex nonlinear relationships between symptoms and the underlying disease.

#### Decision Trees

Three tree-based models:

- Decision Tree
- Random Forest
- Extreme Gradient Boosting (XGBoost)

We applied three tree-based supervised machine learning models --decision tree, random forest, and XGBoost-- in order to predicted the target prognosis using the symptoms and weather features. Using about a 80-20 training and test split, we fit each model on the training data of about 4000 observations. The decision tree model predicted classes by recursively splitting the data based on the most imformative features, but was still very complex after tree pruning. To improve model accuracy and reduce possibilities of overfitting, a random forest model was also fit on the training data. This is an ensemble method that takes the results of multiple decision trees trained on bootstrap samples of data with random feature selection to ensure a robust and generalizable model. Furthermore, we trained a XGBoost model, which is another ensemble method that builds decision trees sequentially using gradient boosting and regularization techniques to reduce overfitting and ensure the loss function is minimized. 

**Best Performance**: XGBoost, $\sim 98.9 \%$ Accuracy

Model performance was evaluated using their respective accuracy, precision, recall, and F1-score on the test data. The XGBoost model achieved the highest classification accuracy over all other models at 98.9%, exhibiting its superior robustness and generalizability.

**Key Predictors**: 

- **Decision Tree**: _chest pain_, _high blood pressure_, _hiv aids_.
- **Random Forest**: _chest pain_, _hiv aids_, _vomiting_.
- **XGBoost**: _chest pain_, _rashes_, _diarrhea_

### Results & Conclusions

**Best Model**: XGBoost ($\sim 98.9 \%$ Accuracy)

- **Neural Networks**: Very strong second ($\sim 95.7 \%$)
- **Regression Models**: Overall lower accuracy ($\sim 83.43 \%$) but revealed the linear relationships.

**Weather Features**: Surprising not significant in predicting prognosis

**Symptom Data**: Were driving basically all the model performance. The models demonstrated that symptoms alone are sufficient to classify disease accurately.

**Conclusion**: Our project shows that machine learning models, especially XGBoost, can classify patient prognosis from symptoms. Weather features, which were thought to contribute to disease, did not make significant contributions in this data set. These findings are encouraging for potential implementations of predictive models in clinical decision support systems to assist in early diagnosis and to instill confidence in patients.

**Future Changes**: We believe that the 3 weather condition variables in the dataset were insufficient for uncovering the relationship between climate and disease. This relationship can be expanded on by accessing hospitals' anonymized symptoms records, e.g. via VitalDB, and merging them with a dataset of more detailed environmental conditions (precipitation, UV index, biome) retrieved from local weather stations' APIs. We could also add data about population density and demographics, level of access to public services, average income, etc. to further explore the biggest contributing factors to disease.
